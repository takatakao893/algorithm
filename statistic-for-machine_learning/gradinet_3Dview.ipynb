```python
import numpy as np
import matplotlib.pyplot as plt

# xy
x = np.arange(-3, 4)
y = np.arange(-3, 4)

# 
X, Y = np.meshgrid(x, y)

# 2
Z = X**2 + Y**2

# ()
dX = 2*X
dY = 2*Y

# xy
fx = np.arange(-3, 3.1, 0.1)
fy = np.arange(-3, 3.1, 0.1)

# 
fX, fY = np.meshgrid(fx, fy)

# 2
fZ = fX**2 + fY**2

# 2D
plt.figure(figsize=(8, 8))
# 
plt.contourf(fX, fY, fZ, alpha=0.5) # 
plt.quiver(X, Y, -dX, -dY)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient', fontsize=20)
plt.grid()
plt.show()

# z
W = np.sqrt(dX**2 + dY**2) + 1e-7
# 3D
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection='3d')
ax.plot_wireframe(X, Y, Z, alpha=0.5, label='f(x, y)')
ax.contourf(fX, fY, fZ, alpha=0.5, offset=0)
ax.quiver(X, Y, Z, -dX/W, -dY/W, -W, color='black', pivot='tail', arrow_length_ratio=0.1, length=0.5, label='(dx, dy)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.legend()
plt.show()
```
